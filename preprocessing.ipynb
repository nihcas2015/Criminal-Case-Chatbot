```python
import os
import shutil

# --- CONFIGURATION ---
source_folder = r"E:\Judgement\dataset\IN-Abs\train-data\New folder"        # Folder with all judgement text files
destination_folder = r"E:\Judgement\crime_cases"      # Folder to save crime-related cases
keywords = [
    "murder", "theft", "robbery", "assault", "fraud", 
    "rape", "kidnapping", "homicide", "burglary", "crime"
]  # You can expand this list

# Create destination folder if it doesn't exist
os.makedirs(destination_folder, exist_ok=True)

# --- PROCESS FILES ---
for filename in os.listdir(source_folder):
    if filename.endswith(".txt"):
        file_path = os.path.join(source_folder, filename)
        with open(file_path, "r", encoding="utf-8") as f:
            text = f.read().lower()  # lowercase for easier matching

        # Check if any crime keyword is present
        if any(keyword in text for keyword in keywords):
            shutil.copy(file_path, os.path.join(destination_folder, filename))
            print(f"Crime-related case found and copied: {filename}")

print("Filtering complete!")
```
