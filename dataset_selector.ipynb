```python
import os
import shutil
from itertools import islice

# --- CONFIGURATION ---
source_folder = r"E:\Judgement\dataset\IN-Abs\train-data\New folder"
destination_folder = r"E:\Judgement\crime_cases"

# Only criminal case indicators in first few lines
case_indicators = [
    "criminal appeal", "state vs", "state v.", "crl.", "criminal revision",
    "charge under ipc", "trial of", "session case", "special case", "ipc"
]

# Create destination folder if it doesn't exist
os.makedirs(destination_folder, exist_ok=True)

# --- PROCESS FILES ---
for filename in os.listdir(source_folder):
    if filename.endswith(".txt"):
        file_path = os.path.join(source_folder, filename)
        with open(file_path, "r", encoding="utf-8") as f:
            # Read up to first 15 lines safely
            lines = [line.lower() for line in islice(f, 15)]
            text_start = " ".join(lines)

        # Strictly check only for criminal indicators
        if any(indicator in text_start for indicator in case_indicators):
            shutil.copy(file_path, os.path.join(destination_folder, filename))
            print(f"Criminal case found and copied: {filename}")

print("Strict criminal case filtering complete!")
```
